# Generador de Presupuestos en PDF

Este proyecto automatiza la creación de un presupuesto formal basado en:
* Descripción del proyecto
* Horas estimadas
* Valor por hora
* Tiempo de entrega

## 1. Importación de Librerías y Configuración

```python
from fpdf import FPDF
import time
```

## 2. Captura de Datos del Cliente

```python
proyecto = input("Escriba una breve descripcion del proyecto: ")
horas_estimadas = int(input("Escribas las horas estimadas para el proyecto: "))
valor_hora = int(input("¿Cual es el valor de la hora trabajada?: "))
tiempo_estimado = input("Escriba cuanto tiempo estima que tomara el proyecto: ")

print("\n--- Datos capturados con éxito ---")
```

## 3. Realización de Cálculos Financieros

```python
valor_total = horas_estimadas * valor_hora

print(f"El valor total calculado es: ${valor_total}")
```

## 4. Generación del Archivo PDF

Se utiliza la librería `FPDF` para posicionar los textos en las coordenadas especificadas.

```python
# Creación del PDF
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)

# El método text(x, y, txt) requiere strings
# Coordenadas ajustadas según tu código original
pdf.text(115, 145, proyecto)
pdf.text(115, 160, str(horas_estimadas))
pdf.text(115, 175, str(valor_hora))
pdf.text(115, 190, tiempo_estimado)
pdf.text(115, 205, str(valor_total))

# Guardar el archivo
nombre_archivo = "Presupuesto.pdf"
pdf.output(nombre_archivo)

print(f"Presupuesto '{nombre_archivo}' generado con éxito")
```

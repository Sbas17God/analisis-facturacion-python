## Buscar datos del precio de las acciones


```python
import yfinance
```

```python
ticker = input("Escriba el codigo de la accion: ")

data = yfinance.Ticker(ticker).history("6mo")
cierre = data.Close
cierre.plot()
```

## Hacer los analisis solicitados por el gestor

* Cotizacion Maxima
* Cotizacion Minima
* Cotizacion Media


```python
maxima = round(cierre.max(), 2)
minima = round(cierre.min(), 2)
Valor_Medio = round(cierre.mean(), 2)
```

## Mandar un correro electronico

```python
import webbrowser
import pyautogui
import pyperclip
import time
```

```python
pyperclip.set_clipboard("pbcopy")

webbrowser.open("https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox")

destinatario = "sebasmoreno04@gmail.com"
asunto = "Analisis de acciones de los ultimos 6 meses"

mensaje = f"""
Buenas noches Sebastian.

Te mando el analisis de las acciones de {ticker} en los ultimos 6 meses

Cotizacion Maxima: ${maxima} usd
Cotizacion Minima: ${minima} usd
Cotizacion Media:  ${Valor_Medio} usd

Quedo pendiente a cualquier observacion!
"""

print("Destinatario:", destinatario)

time.sleep(10)

pyautogui.PAUSE = 3
pyautogui.click(114, 184)

pyperclip.copy(destinatario)
time.sleep(3)  
pyautogui.hotkey("cmd", "v")
pyautogui.hotkey("tab")

print("Asunto:", asunto)

pyperclip.copy(asunto)
pyautogui.hotkey("cmd", "v")

pyautogui.hotkey("tab")
```
